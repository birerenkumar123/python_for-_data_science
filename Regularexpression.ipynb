```python

'''Input: s = "aa", p = "a"
Output: false
Explanation: "a" does not match the entire string "aa".
Example 2:

Input: s = "aa", p = "a*"
Output: true
Explanation: '*' means zero or more of the preceding element, 'a'. Therefore, by repeating 'a' once, it becomes "aa".
Example 3:

Input: s = "ab", p = ".*"
Output: true
Explanation: ".*" means "zero or more (*) of any character (.)".'''
 
```

```python
def ismatch(p, s):
    # Base case: if pattern is empty, string must be empty
    if not p:
        return not s
    # Check if first character matches (or is '.')
    first_match = bool(s) and (p[0] == s[0] or p[0] == '.')
    # If the next character is '*', handle zero or more occurrences
    if len(p) >= 2 and p[1] == '*':
        # Zero occurrences: skip the "x*" in pattern
        # One or more: if first matches, recurse on s[1:] with same pattern
        return ismatch(p[2:], s) or (first_match and ismatch(p, s[1:]))
    else:
        # No '*', just check first and recurse
        return first_match and ismatch(p[1:], s[1:])

ismatch('.','a')
```
