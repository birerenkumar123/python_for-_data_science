```python
def ispallindrome(ss):
    i=0
    j=len(ss)-1
    while i<=j:
        if ss[i]!=ss[j]:
            return False
        else:
            i+=1
            j-=1
    return True
            
        
def pallindromic_substring(s):
    for i in range(0,len(s)):
        for j in range(0,len(s)+1):
            ss=s[i:j]
            if ispallindrome(ss):
                print(ss)
                
```

```python
def my_upper(s):
    result=''
    for c in s:
        if 'a'<=c<='z':
            s1=chr(ord(c)-32)
            result+=s1
    return result
print(my_upper("mayur"))
```

```python
def my_lower(s):
    result=''
    for c in s:
        if 'A'<=c<='Z':
            s1=chr(ord(c)+32)
            result+=s1
    return result
print(my_lower("MAYUR"))
```

```python
def string_compression(s):
    result=""
    i=0
    j=i+1
    while i<len(s)-1:
        if s[i]!=s[j]:
            result+=s[i]
        i+=1
        j+=1
    return result+s[len(s)-1]
s="aaaaabbbbbbcccccc"
print(string_compression(s))
```

```python
#o/p:AbCdEf
#s="aBcDeF"
def case_conversion(s):
    result=""
    for i in range(len(s)):
        if "a"<=s[i]<="z":
            var=chr(ord(s[i])-32)
            result+=var
            
        elif "A"<=s[i]<="Z":
            var=chr(ord(s[i])+32)
            result+=var
        else:
            result+=s[i]
    return result
s="aBcDeF"   
print(case_conversion(s))
```

```python
#S="abcdefgshgfgkdkdlslshdhd"
def return_vowel_consonant_dict(s):
    result={}
    vowel=0
    consonant=0
    for i in s:
        if i in ["a","e","i","o","u"]:
            vowel+=1
        else:
            consonant+=1
    result["vowel"]=vowel
    result["consonant"]=consonant
    return result
S="abcdefgshgfgkdkdlslshdhd"
print(return_vowel_consonant_dict(S))
            
```

```python
def max_length_word_in_string(ss):
    lst=ss.split(' ')
    max1=len(lst[0])
    for i in range(1,len(lst)):
        if max1<len(lst[i]):
            max1=lst[i]
    return max1
ss='''this is the my approach for coding'''
print(max_length_word_in_string(ss))
```

```python
def anagram(s1,s2):
    for i in s1:
        if i in s2:
            return True
        else:
            return False
s1="listen"            
s2="s","i","l","e","n","t"
print(anagram(s1,s2))
```

```python
def anagram_check(s1,s2):
    yes=0
    no=0
    s3=list(s1)
    s4=list(s2)
    if len(s1)!=len(s2):
        return False
    for i in s3:
        if i in s4:
            yes+=1
        else:
            no+=1
    if no>0:
        return False
    else:
        return True
s1="silent"
s2="listen"
print(anagram_check(s1,s2))
```

```python
'''remove sub string:"aaabbb" → "ab" '''
def remove_duplicates(s):
    result = []
    for char in s:
        if not result or char != result[-1]:
            result.append(char)
    return ''.join(result)


s = "aaabbb"
print(remove_duplicates(s))  # Output: "ab"




```
