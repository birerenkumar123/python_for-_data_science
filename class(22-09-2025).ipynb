```python
#how many 7 in present in the list
def my_count(lst,element):
    count=0
    for i in lst:
        if element==i:
            count+=1
    return count
lst=[3,4,5,6,77,7,8,9,10,7,7,9]
print(my_count(lst,7))
```

```python
def count_recursive(lst,element):
    if lst==[]:
        return 0
    elif lst[0]==element:
        return 1+count_recursive(lst[1:],element)
    else:
        return count_recursive(lst[1:],element)
lst=[3,4,5,6,77,7,8,9,10,7,7,9]
print(count_recursive(lst,7))
```

```python
def my_sum(lst,cnt):
    for i in lst:
        cnt+=i
    return cnt
lst = [1,2,3,4,5]
my_sum(lst,0)
```

```python
def my_sum_recursive(lst):
    if lst==[]:
        return 0
    else:
        return lst[0]+my_sum_recursive(lst[1:])
lst = [1,2,3,4,5]
print(my_sum_recursive(lst))
        
```

```python
#any iteration use
def any(t):
    for i in t:
        if t!=0:
            return True
    return False
             
    
t=(0,0,0,0) 
print(any(t))
```

```python
#any recursive
def any(t):
    i=0
    if t[i]!=0:
        return True
    else:
        return False
    i+=1

any((0,0,0,0))
```

```python
#length define in tuple
def len1(t):
    if t==[]:
        return 0
    else:
        return 1+ len(t[1:])

len1((1,2,2,2,2,2,2,2,2,2,2,2))
```

```python
#min
numbers = [10, 3, 5, 8, 2]

# Assume the first number is the minimum
minimum = numbers[0]

for num in numbers:
    if num < minimum:
        minimum = num

print("Minimum value:", minimum)
```
