```python
def sorting_str(st):
    result=""
    for i in range(len(st)):
        count=1
        if st[i] not in result:
            for j in range(i+1,len(st)):
                if st[i]==st[j]:
                    count+=1
            while count!=0:
                result+=st[i]
                count-=1
    return result
st="ababcdabcbcbabba"
sorting_str(st)
```

```python
def same_string(s):
    d1 = {}
    for i in s:
        if i in d1:
            d1[i] += 1
        else:
            d1[i] = 1
    return d1
print(same_string("ababcccdbba"))
```

```python
def shifting_all_zeros_right(lst):
    i=0
    j=len(lst)-1
    while i<j:
        if lst[i]==0 and lst[j]>0:
            lst[i],lst[j]=lst[j],lst[i]
            i+=1
            j-=1
        elif lst[i]>0 and lst[j]==0:
            j-=1
        elif lst[i]>0 and lst[j]>0:
            i+=1
        else:
            if lst[j]==0:
                while lst[j]>0:
                    j-=1
                lst[i],lst[j]=lst[j],lst[i]
    return lst
lst=[2,3,0,4,5,0,7,0,5,0,0,9]                
print(shifting_all_zeros_right(lst))            
        
        
        
            
        
```
