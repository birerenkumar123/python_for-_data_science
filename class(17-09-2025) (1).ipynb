```python
def finding_peak(lst):
    for i in range(1,len(lst)):
        if lst[i-1]<lst[i] and lst[i+1]<lst[i]:
            print(lst[i])
lst=[2,3,1,4,5,3,7,8,9,-3,11,2]
print(finding_peak(lst))
```

```python
def finding_peak_change(lst):
    for i in range(1,len(lst)-1):
        if lst[i-1]<lst[i] and lst[i+1]<lst[i]:
            print(lst[i])
    if lst[0]>lst[1]:
        print(lst[0])
    if lst[-1]>lst[-2]:
        print(lst[-1])
lst=[12,2,3,1,4,5,3,7,8,9,-3,11,2,16]
print(finding_peak_change(lst))
```

```python
def ishan_loves_chocalates(lst):
    i=0
    j=len(lst)-1
    while i!=j:
        if lst[i]<lst[j]:
            j-=1
        else:
            i+=1
    return lst[j]
lst = [5, 3, 1, 6, 9]
print(ishan_loves_chocalates(lst))
```

```python
def bheem_laddoo(lst):
    lst.sort()
    return lst[0]

lst= [5, 3, 1, 6, 9]
bheem_laddoo(lst)
```

```python
def replacing_null(lst):
    for i in range(1,len(lst)):
        if lst[i]=='null':
            lst[i]=lst[i-1]
    return lst


lst=[12, 14, 'null','null', 7,'null', 19, 'null', 'null', 23]
print(replacing_null(lst))
```

```python
#recursion 
def di(n):
    if n==0:
        return
    di(n-1)
    print(n)

di(5)
```
