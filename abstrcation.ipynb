```python
from abc import ABC,abstractmethod
class spotify(ABC):
    def display(self):
        print("please implement the abstract method")
    @abstractmethod
    def play(self):
        pass
    @abstractmethod
    def pause(self):
        pass
class song(spotify):
    def own_song(self):
        print("humari adhuri kahanai")
    def play(self):
        print("playing a song")
    def pause(self):
        print("pause the song")
    
```

```python
s=song()
```

```python
s.own_song()
```

```python
s.play()
```

```python
s.pause()
```

```python
s.display()
```
