```python
def create_LL(lst):
    if len(lst)==0:
        return None
    start={"data":lst[0],"link":None}
    ptr=start
    for i in range(1,len(lst)):
        new_node={"data":lst[i],"link":None}
        ptr["link"]=new_node
        ptr=new_node
    return start
lst=[2,3,4,5,6,7,8,9]
print(create_LL(lst))
```

```python
def insert_at_end(start,node):
    new_node={"data":node,"link":None}
    ptr=start
    while ptr["link"]!=None:
        ptr=ptr["link"]
    ptr["link"]=new_node
    return start
start={'data': 2, 'link': {'data': 3, 'link': {'data': 4, 'link': {'data': 5, 'link': {'data': 6, 'link': {'data': 7, 'link': {'data': 8, 'link': {'data': 9, 'link': None}}}}}}}}
print(insert_at_end(start,1000))
```

```python
def insert_in_between(start,node,pos):
    new_node={"data":node,"link":None}
    ptr=start
    for i in range(0,(pos-1)):
        ptr=ptr["link"]
    tmp=ptr["link"]
    ptr["link"]=new_node
    ptr=new_node
    ptr["link"]=tmp
    return start
start={'data': 2, 'link': {'data': 3, 'link': {'data': 4, 'link': {'data': 5, 'link': {'data': 6, 'link': {'data': 7, 'link': {'data': 8, 'link': {'data': 9, 'link': {'data': 1000, 'link': None}}}}}}}}}
print(insert_in_between(start,345,3))
    
    
```
