```python
def two_sum(lst,target):
    for i in range(0,len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i]+lst[j]==target:
                return i,j
nums = [2,7,11,15]
target = 9
print(two_sum(nums,target))

                
```

```python
def two_sum(lst,target):
    dic={}
    j=0
    for i in lst:
        dic[i]=j
        j+=1
    for k in dic:
        for l in dic:
            if k+l==target:
                return dic[k],dic[l]
lst=[2,7,8,0,5]
target=9
two_sum(lst,target)
```

```python
def two_sum_optimized(lst,target):
    dict1={}
    complement=0
    for i in range(0,len(lst)):
        dict1[lst[i]]=i
    for i in range(0,len(lst)):
        complement=target-lst[i]
        if complement in dict1:
            return i,dict1[complement]


lst=[2,7,11,15]  
target=9
print(two_sum_optimized(lst,target))    
```

```python
def three_sum_optimized(lst,target):
    small={}
    complement=0
    for i in range(0,len(lst)):
        small[lst[i]]=i
    for j in range(0,len(lst)):
        for k in range(j+1,len(lst)):
            complement=target-(lst[j]+lst[k])
            if complement in small:
                return j,k,small[complement]
lst=[2,3,4,7,9]
target=9
print(three_sum_optimized(lst,target))   
```
