```python
def recursive_multiplication(a,b):
    if b==0:
        return 0
    if b>0:
        a+=recursive_multiplication(a,(b-1))
    else:
        a+=recursive_multiplication(a,(b+1))
    if b%2==0:
        return a
    else:
        return -a
a=int(input("enter any numer"))
b=int(input("enter any numer"))
print(recursive_multiplication(a,b))
```

```python
def rec_mul(a,b):
    if b==0:
        return 0
    elif b<0:
        return -a+rec_mul(a,(b+1))
    else:
        return a+rec_mul(a,(b-1))
a=int(input("enter any numer"))
b=int(input("enter any numer"))
print(rec_mul(a,b))
```

```python
def recursive_division(a,b):
    if a==0:
        return 0
    elif a<0:
        return 1+recursive_division((a+b),b)
    elif a<0 and b<0:
        return -1-recursive_division((a+b),b)
    else:
        return 1+recursive_division((a-b),b)
a=int(input("enter any numer"))
b=int(input("enter any numer"))
print(recursive_division(a,b))
```

```python
def power(a,n):
    if n==0:
        return 1
    else:
        return a*power(a,(n-1))
a=int(input("enter any numer"))
n=int(input("enter any numer"))
print(power(a,n))
```
