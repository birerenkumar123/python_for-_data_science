```python
#WAP to count the number of alphabets,digits and special characters
f14=open("xyz.txt","r")
string=f14.read()
a=0
d=0
sc=0
for ch in string:
    if ch.isalpha():
        a+=1
    elif ch.isdigit():
        d+=1
    else:
        sc+=1
print("no of alphabets:",a)
print("no of digits:",d)
print("no of Special characters:",sc)
f14.close()
```
