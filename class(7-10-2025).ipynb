```python
def merge_dictionaries(dict1,dict2):
    merge={}
    for i in dict1:
        merge[i]=dict1[i]
  
    for j in dict2:
        merge[j]=dict2[j]
    return merge
dict1={
    1:101,
    2:201,
    3:301
}
dict2={
    4:401,
    5:501,
    6:601
}
merge_dictionaries(dict1,dict2)
```

```python
def sort_dict(d):
    d1 = {}
    items = list(d.items())
    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            if items[i][1] > items[j][1]:
                items[i], items[j] = items[j], items[i]
    for k, v in items:
        d1[k] = v
    return d1
d = {4: 12, 3: 13, 2: 14, 1: 15}
print(sort_dict(d))
    
    
    
```

```python
graph={}
def creation_graph(v1,v2):
    
    if v1 not in graph:
        graph[v1]=[]
    if v2 not in graph:
        graph[v2]=[]
    graph[v1].append(v2)
    graph[v2].append(v1)
    return graph
    
        
```

```python
creation_graph('A','B')
```

```python
creation_graph('A','C')
```

```python
creation_graph('B','D')
```

```python
graph={}
def directed_graph(v1,v2):
    if v1 not in graph:
        graph[v1]=[]
    if v2 not in graph:
        graph[v2]=[]
    graph[v1].append(v2)
    return graph
```
