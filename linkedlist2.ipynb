```python
def create_LL(lst):
    if len(lst)==0:
        return None
    start={"data":lst[0],"link":None}
    ptr=start
    for i in range(1,len(lst)):
        new_node={"data":lst[i],"link":None}
        ptr["link"]=new_node
        ptr=ptr["link"]
    return start
lst=[2,3,4,5,6,7,8]
create_LL(lst)
```

```python
def count_no_nodes(start):
    count=0
    ptr=start
    while ptr!=None:
        count+=1
        ptr=ptr["link"]
    return count
start={'data': 2,
 'link': {'data': 3,
  'link': {'data': 4,
   'link': {'data': 5,
    'link': {'data': 6, 'link': {'data': 7, 'link': None}}}}}}
count_no_nodes(start)
```

```python
def delete_kth_node_from_left(start,k):
    n=count_no_nodes(start)
    pos=n-k
    ptr=start
    for i in range(0,pos-1):
        ptr=ptr["link"]
    tmp=ptr["link"]
    while ptr!=None or ptr["link"]!=None:
        ptr["link"]=ptr["link"]["link"]
        break
    return start
start={'data': 2,
 'link': {'data': 3,
  'link': {'data': 4,
   'link': {'data': 5,
    'link': {'data': 6, 'link': {'data': 7, 'link': None}}}}}}
delete_kth_node_from_left(start,3)    
    
```

```python
def get_middle_element_LL(start):
    slow=fast=start
    while fast!=None and fast["link"]!=None:
        slow=slow["link"]
        fast=fast["link"]["link"]
    return slow["data"]
start={'data': 2,
 'link': {'data': 3,
  'link': {'data': 4,
   'link': {'data': 5,
    'link': {'data': 6,
     'link': {'data': 7, 'link': {'data': 8, 'link': None}}}}}}}
get_middle_element_LL(start)    
```

```python
def detect_loop_LL(start):
    slow=fast=start
    while fast!=None and fast["link"]!=None:
        slow=slow["link"]
        fast=fast["link"]["link"]
        if slow is fast:
            return True
    return False
start={"data":2,"link":None}
node2={"data":3,"link":None}
node3={"data":4,"link":None}
node4={"data":5,"link":None}
node5={"data":6,"link":None}
start["link"]=node2
node2["link"]=node3
node3["link"]=node4
node4["link"]=node5
node5["link"]=node4
detect_loop_LL(start)
    
```

```python
def reverse_LL(start):
    curr=start
    prev=None
    while curr!=None:
        tmp=curr["link"]
        curr["link"]=prev
        prev=curr
        curr=tmp
    return prev
start={'data': 2,
 'link': {'data': 3,
  'link': {'data': 4,
   'link': {'data': 5,
    'link': {'data': 6,
     'link': {'data': 7, 'link': {'data': 8, 'link': None}}}}}}}   
reverse_LL(start)
```
