```python
def creating_dict_adding_values(d1,d2):
    result={}
    for i in d1:
        for j in d2:
            if i==j:
                result[i]=d1[i]+d2[j]
    return result
        
```

```python
d1={
   1:100,
   2:200,
   3:300
   }
d2={
   1:400,
   2:600,
   3:700
   }
print(creating_dict_adding_values(d1,d2))
```

```python
def create_LL(lst):
    if len(lst)==0:
        return None
    start={"data":lst[0],"link":None}
    ptr=start
    for i in range(1,len(lst)):
        new_node={"data":lst[i],"link":None}
        ptr["link"]=new_node
        ptr=new_node
    return start
        
        
    
    
```

```python
lst=[2,3,4,5,6,7,8]
print(create_LL(lst))
```

```python
def count_nodes_LL(start):
    ptr=start
    count=0
    while ptr!=None:
        count+=1
        ptr=ptr["link"]
    return count
start={'data': 2, 'link': {'data': 3, 'link': {'data': 4, 'link': {'data': 5, 'link': {'data': 6, 'link': {'data': 7, 'link': {'data': 8, 'link': None}}}}}}} 
print(count_nodes_LL(start))
```

```python
def linear_search_LL(start,key):
    ptr=start
    while ptr!=None:
        if ptr["data"]==key:
            return True
        ptr=ptr["link"]
    return False
start={'data': 2, 'link': {'data': 3, 'link': {'data': 4, 'link': {'data': 5, 'link': {'data': 6, 'link': {'data': 7, 'link': {'data': 8, 'link': None}}}}}}} 
print(linear_search_LL(start,7))
```
