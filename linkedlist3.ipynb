```python
def create_DLL(lst):
    if len(lst)==0:
        return None
    start={"data":lst[0],"prev":None,"next":None}
    ptr=start
    for i in range(1,len(lst)):
        new_node={"data":lst[i],"prev":ptr,"next":None}
        ptr["next"]=new_node
        ptr=new_node
    return start
lst=[2,3,4,5,6,7,8,9]
create_DLL(lst)
```

```python
def insertion_at_end(start,data):
    ptr=start
    new_node={"data":data,"prev":None,"next":None}
    while ptr["next"]!=None:
        ptr=ptr["next"]
    ptr["next"]=new_node
    new_node=ptr["prev"]
    return start
start={'data': 2,
 'prev': None,
 'next': {'data': 3,
  'prev': {...},
  'next': {'data': 4,
   'prev': {...},
   'next': {'data': 5,
    'prev': {...},
    'next': {'data': 6,
     'prev': {...},
     'next': {'data': 7,
      'prev': {...},
      'next': {'data': 8,
       'prev': {...},
       'next': {'data': 9, 'prev': {...}, 'next': None}}}}}}}}
insertion_at_end(start,200)
    
```

```python
def insertion_in_between(start,data,pos):
    ptr=start
    new_node={"data":data,"prev":None,"next":None}
   
    for i in range(0,pos-1):
        ptr=ptr["next"]
    tmp=ptr["next"]
    ptr["next"]=new_node
    ptr=new_node
    new_node["next"]=tmp
    new_node["prev"]=ptr
    return start
start={'data': 2,
 'prev': None,
 'next': {'data': 3,
  'prev': {...},
  'next': {'data': 4,
   'prev': {...},
   'next': {'data': 5,
    'prev': {...},
    'next': {'data': 6,
     'prev': {...},
     'next': {'data': 7,
      'prev': {...},
      'next': {'data': 8,
       'prev': {...},
       'next': {'data': 9, 'prev': {...}, 'next': None}}}}}}}}
insertion_in_between(start,456,4)        
    
    
```
