```python
def my_length(s):
    count=0
    for i in s:
        count+=1
    return count
s={1, 2, 3, 4, 5, 6, 7, 8, 9, 1000}
print(my_length(s))
```

```python
def recursive_my_length(s):
    if not s:
        return 0
    else:
        s.pop()
        return 1+recursive_my_length(s)
s={1, 2, 3, 4, 5, 6, 7, 8, 9, 1000}
print(recursive_my_length(s))
```

```python
def my_discard(s1,ele):
    result=set()
    for i in s1:
        if ele!=i:
            result.add(i)
    return result
s={1, 2, 3, 5, 6, 7, 8, 9, 1000}
print(my_discard(s1,1000))
            
        
```

```python
def my_discard_recursive(s,ele,result=set()):
    
    if not s:
        return result
    else:
        tmp=s.pop()
       
        if tmp!=ele:
            result.add(tmp)
            
        return my_discard_recursive(s,ele)
s={1, 2, 3, 5, 6, 7, 8, 9, 1000}

print(my_discard_recursive(s,1000))
    
    
```

```python
def union_rec(s1,s2):
    if not s2:
        return s1
    temp = s2.pop()
    s1.add(temp)
    return union_rec(s1,s2)
s1={1,2,3,4}
s2={5,6,7}
print(union_rec(s1,s2))
    
    
```

```python
def my_set_intersection(s1,s2):
    result=set()
    for i in s1:
        for j in s2:
            if i==j:
                result.add(i)
    return result
s1={2,3,4,5,6,7,8,9}
s2={5,6,7,8,9,10}
print(my_set_intersection(s1,s2))
```

```python
def intersection_rec(s1,s2,result=set()):
    if not s2:
        return result
    else:
        temp=s2.pop()
        if temp in s1:
            result.add(temp)
        return intersection_rec(s1,s2,result)
s1={2,3,4,5,6,7,8,9}
s2={5,6,7,8,9,10}
print(intersection_rec(s1,s2))
```

```python
def set_differenec(s1,s2):
    result=set()
    for i in s1:
        if i not in s2:
            result.add(i)
    return result
s1={2,3,4,5,6,7,8,9}
s2={5,6,7,8,9,10}
print(set_differenec(s1,s2))
```

```python
def set_diff_rec(s1,s2):
    if not s2:
        return s1
    else:
        temp=s2.pop()
        if temp in s1:
            s1.discard(temp)
        return set_diff_rec(s1,s2)
s1={2,3,4,5,6}
s2={5,6,7,8,9}
set_diff_rec(s1,s2)
```

```python
def my_symetric_diff(A, B):
    result = set()
    for i in A:
        if i not in B:
            result.add(i)
    for i in B:
        if i not in A:
            result.add(i)
    return result
A = {1, 2, 3,4, 6, 8}
B = {3, 4, 5, 6, 9}
print(my_symetric_diff(A, B))
```

```python
def sym_diff_rec(s1,s2):
    if not s2:
        return s1
    else:
        temp=s2.pop()
        if temp in s1:
            s1.discard(temp)
        else:
            s1.add(temp)
        return sym_diff_rec(s1,s2)
s1={2,3,4,5,6}
s2={5,6,7,8,9}
sym_diff_rec(s1,s2)
```
