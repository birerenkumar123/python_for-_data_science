```python
class stack:
    def __init__(self):
        self.lst=[]
    def push(self,element):
        self.lst.append(element)
        return self.lst
    def pop(self):
        if self.isempty():
            print("stack is empty")
        else:
            self.lst.pop()
            return self.lst
    def isempty(self):
        if len(self.lst)==0:
            return True
        else:
            return False
    def peek(self):
        if self.isempty():
            print("stack is empty")
        else:
            return self.lst[-1]
s=stack()
s.push(344)
```

```python
class queue:
    def __init__(self):
        self.lst=[]
    def enqueue(self,element):
        self.lst.append(element)
        return self.lst
    def dequeue(self):
        if self.isempty():
            print("queue is empty")
        else:
            self.lst.pop(0)
            return self.lst
    def isempty(self):
        if len(self.lst)==0:
            return True
        else:
            return False
    def peek(self):
        if len(self.lst)==0:
            return True
        else:
            return self.lst[-1],self.lst

q=queue()
```

```python
q.enqueue(1)
```

```python
class Stack:
    def __init__(self):
        self.lst = []

    def push(self, ele):
        self.lst.append(ele)
        return self.lst

    def pop(self):
        if self.isEmpty():
            print("Stack is empty")
        else:
            return self.lst.pop()

    def isEmpty(self):
        return len(self.lst) == 0

    def peek(self):
        if self.isEmpty():
            print("Stack is empty")
        else:
            return self.lst[-1]

    def get_min(self):
        if self.isEmpty():
            print("Stack is empty")
        else:
            current_min = self.lst[0]
            for i in self.lst:
                if i < current_min:
                    current_min = i
            return current_min
s=Stack()
```

```python
s.push(7)
s.push(70)
```

```python
s.pop()
```

```python
s.peek()
```

```python
class student:
    def __init__(self,rollno,name):
        self.rollno=rollno
        self.name=name
    
class graduate(student):
    def __init__(self,rollno,name,degree):
        super().__init__(rollno,name)
        self.degree=degree
    def display(self):
        print(self.rollno)
        print(self.name)
        print(self.degree)
```

```python
class student:
    def __init__(self,roll,name):
        self.roll = roll
        self.name = name
    def marks(self,marks):
        self.marks =  marks
class graduate(student):
    def __init__(self,roll,name,degree,marks):
        super().__init__(roll,name)
        super().marks(marks)
        self.degree = degree
    def display(self):
        print(self.roll)
        print(self.name)
        print(self.degree)
        print(self.marks)
```

```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
class student(person):
    def __init__(self,name,age,roll,grades):
        super().__init__(name,age)
        self.roll = roll
        self.grades = grades
class graduate(student):
    def __init__(self,name,age,roll,grades,degree):
        super().__init__(name,age,roll,grades)
        self.degree = degree
    def display(self):
        print(self.name)
        print(self.age)
        print(self.roll)
        print(self.grades)
        print(self.degree)
```

```python
class Vehical:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
class Car(Vehical):
    def __init__(self,brand,model,number,doors):
        super().__init__(brand,model)
        self.number = number
        self.doors = doors
class ElectricCar (Car):
    def __init__(self,brand,model,number,doors,battery,capacity):
        super().__init__(brand,model,number,doors)
        self.battery = battery
        self.capacity = capacity
    def display(self):
        print(self.brand)
        print(self.model)
        print(self.number)
        print(self.doors)
        print(self.battery)
        print(self.capacity)
```
