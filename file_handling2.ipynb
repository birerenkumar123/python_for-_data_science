```python
fd=open("tp.txt","r")
lst=fd.readlines()
for line in lst:
    data=line.split()
    largest=data[0]
    for word in data:
        if len(word)>len(largest):
            largest=word
    print(largest)
fd.close()
        
    
    
```

```python
#WAP to count the number of vowels and consonant in each line
fd=open("tp.txt","r")
data=fd.read()
lst=data.split()

for sen in lst:
    vowel=0
    cons=0
    for ch in sen:
        if ch in ["a","e","i","o","u"]:
            vowel+=1
        else:
            cons+=1
    print("vowel:",vowel)
    print("consonant:",cons)
fd.close()
```

```python
f1=open("tp.txt","r")
data=f1.read()
lst=data.split()
largest=lst[0]
smallest=lst[0]

for i in range(0,len(lst)):
   
    if len(lst[i])>len(largest):
        largest=lst[i]
for i in range(0,len(lst)):
   
    if len(lst[i])<len(smallest):
        smallest=lst[i]
f2=open("tp3.txt","a")
f2.write(largest)
f2.write(smallest)


f1.close()
f2.close()
```

```python
f4=open("tp4.txt","r")
data=f4.read()
f5=open("tp5.txt","w")
for c in data:
    if c.isalpha():
        f5.write("*")
    elif c.isdigit():
        f5.write("#")
    else:
        f5.write("-")
f4.close()
f5.close()

        
        
```
