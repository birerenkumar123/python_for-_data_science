```python
class undirectd_graph:
    def __init__(self):
        self.graph={}
    def add_vertex(self,vertex):
        if vertex not in self.graph:
            self.graph[vertex]=[]
    def add_edge(self,vertex1,vertex2):
        if vertex1 not in self.graph:
            self.add_vertex(vertex1)
        if vertex2 not in self.graph:
            self.add_vertex(vertex2)
        self.graph[vertex1].append(vertex2)
        self.graph[vertex2].append(vertex1)
    def display_graph(self):
        print(self.graph)
```

```python
g.add_edge("B","C")
```

```python
g.add_edge("C","D")
```

```python
g.add_edge("D","E")
```

```python
g.display_graph()
```

```python
class Directed_Graph:
    def __init__(self):
        self.graph={}
    def add_edge(self,vertex1,vertex2):
        if vertex1 not in self.graph:
            self.graph[vertex1]=[]
        if vertex2 not in self.graph:
            self.graph[vertex2]=[]
        self.graph[vertex1].append(vertex2)
    def display_graph(self):
        print(self.graph)
```

```python
class weighted_graph:
    def __init__(self):
        self.graph={}
    def add_vertex(self,vertex):
        if vertex not in self.graph:
            self.graph[vertex]=[]
    def add_edge(self,vertex1,vertex2,weight):
        if vertex1 not in self.graph:
            self.add_vertex(vertex1)
        if vertex2 not in self.graph:
            self.add_vertex(vertex2)
        self.graph[vertex1].append((vertex2,weight))
        self.graph[vertex2].append((vertex1,weight))
    def display_graph(self):
        print(self.graph)
    
```

```python
wg=weighted_graph()
```

```python
wg.add_edge("B","C",8)
```

```python
wg.add_edge("C","D",10)
```

```python
wg.add_edge("D","F",13)
```

```python
wg.display_graph()
```

```python
def count_vertices_edges(graph):
    vertices=len(graph)
    edges=set()
    for key,values in graph.items():
        for i in values:
            st=""
            st+=key
            st+=i
            edges.add("".join(sorted(st)))
    return vertices,len(edges)
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B', 'D'], 'D': ['C', 'E'], 'E': ['D']}
count_vertices_edges(graph)
```

```python
def vertex_and_edge(graph):
    n=len(graph)
    total=0
    for i in graph:
        total+=len(graph[i])
    return (n,total//2)
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B', 'D'], 'D': ['C', 'E'], 'E': ['D']}
vertex_and_edge(graph)
```

```python
def count_vertices_edges_corrected_v1(graph):
    vertices = len(graph)
    edges = set()
    for key, values in graph.items():
        for i in values:
            st = ""
            st = (key + i)
            edges.add(tuple(sorted(st)))  
    return vertices, len(edges)
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B', 'D'], 'D': ['C', 'E'], 'E': ['D']}
count_vertices_edges_corrected_v1(graph)
```

```python
def get_min_weight_edge(graph):
    res=[]
    for k,v in graph.items():
        for i in v:
            res.append(i[1])
    return min(res)
graph={'A': [('B', 3)], 'B': [('A', 3), ('C', 8)], 'C': [('B', 8), ('D', 10)], 'D': [('C', 10), ('F', 13)], 'F': [('D', 13)]}
get_min_weight_edge(graph)
```

```python
def degree_vertex_graph(graph,vertex):
    if vertex in graph:
        return len(graph[vertex])
    else:
        return 0
graph={'A': ['B'], 'B': ['A', 'C'], 'C': ['B', 'D'], 'D': ['C', 'E'], 'E': ['D']}
degree_vertex_graph(graph,"C")
```

```python
def indegree_outdegree_vertices(graph,vertex):
    in_deg=0
    out_deg=0
    for i,v in graph.items():
        if i==vertex:
            out_deg=len(v)
        if vertex in v:
            in_deg+=1
    return (in_deg,out_deg)
graph={'A': ['B'], 'B': ['E'], 'C': ['D'], 'D': ['E'], 'E': []}   
indegree_outdegree_vertices(graph,'E')   
```
