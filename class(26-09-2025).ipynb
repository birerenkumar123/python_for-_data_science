```python
def fibo(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
print(fibo(5))
```

```python
def climbing_starircase(n):
    if n==1:
        return 1
    elif n==0:
        return 1
    else:
        return climbing_starircase(n-1)+climbing_starircase(n-2)
print(climbing_starircase(2))
```

```python
def reverse_lst(lst):
    if lst==[]:
        return []
    elif len(lst)==1:
        return lst
    else:
        last_ele=lst.pop()
        return [last_ele]+reverse_lst(lst)

lst=[12,11,7,19,21,25]
print(reverse_lst(lst))
```
