```python
def power(a,n):
    if n==0:
        return 1
    else:
        return a*power(a,(n-1))
print(power(5,3))
```

```python
def variation(a,n):
    if n==0:
        return 1
    elif n%2==0:
        res=variation(a,n//2)
        return res*res
    else:
        res=variation(a,n//2)
        return res*res*a
print(variation(5,5))
```

```python
def sum_of_n_elements(n):
    if n==1:
        return 1
    else:
        return n+sum_of_n_elements(n-1)
print(sum_of_n_elements(10))
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
def my_length(lst):
    count=0
    for i in lst:
        count+=1
    return count
lst=[2,3,4,5,6,7,8,9]
print(my_length(lst))
```

```python
def recursive_length(lst):
    if lst==[]:
        return 0
    else:
        return 1+recursive_length(lst[1:])
lst=[2,3,4,5,6,7,8,9]
print(recursive_length(lst))
        
```

```python
def my_min(lst):
    min_ele=lst[0]
    max_ele=lst[0]
    for i in lst:
        if i<min_ele:
            min_ele=i
    for i in lst:
        if i>max_ele:
            max_ele=i
    return 'min:',min_ele,'max:',max_ele
lst=[2,-9,-7,111,23,98,1,-99,-2,3,4,1111]
print(my_min(lst))
    
```
