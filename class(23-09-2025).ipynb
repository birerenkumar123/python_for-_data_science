```python
def ishan_loves_chocalates(lst):
    i=0
    j=len(lst)-1
    while i!=j:
        if lst[i]<lst[j]:
            j-=1
        else:
            i+=1
    return lst[i]
lst = [5, 3, 1, 6, 9]
print(ishan_loves_chocalates(lst))
```

```python
def max_product(lst):
    lst.sort()
    return lst[-1]*lst[-2]
lst= [1, 4, 3, 6, 7, 0] 
 
print(max_product(lst))
```

```python
def matrix_addition(matrix1,matrix2):
    result=[[0,0],[0,0]]
    for i in range(0,len(matrix1)):
        for j in range(0,len(matrix1[0])):
            result[i][j]=matrix1[i][j]+matrix2[i][j]
    return result
matrix1=[[1,2],[3,4]]
matrix2=[[4,5],[6,7]]
print(matrix_addition(matrix1,matrix2))
            
```

```python
def trace_matrix(matrix):
    trace=0
    for i in range(0,len(matrix)):
        for j in range(0,len(matrix[0])):
            if i==j:
                trace+=matrix[i][j]
    return trace
matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(trace_matrix(matrix))
    
    
```

```python
def searching(matrix,key):
    for i in range(0,len(matrix)):
        for j in range(0,len(matrix[0])):
            if matrix[i][j]==key:
                return True
    return False
matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(searching(matrix,8))
```

```python
def matrix_sort(matrix):
    tmp=0
    for i in range(0,len(matrix)):
        for j in range(0,len(matrix[0])):
            if matrix[i][j]>matrix[i][j+1]:
                tmp=matrix[i][j]
                matrix[i][j+1]=matrix[i][j+1]
                matrix[i][j+1]=tmp
    return matrix
matrix=[[-2,1,0],[3,-1,5],[0,4,-4]]
print(matrix_sort(matrix))
            
```
