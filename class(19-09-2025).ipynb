```python
lst1=[2,3,4,5,6,7,8]
lst2=[4,5,6,7,8,9,10]
```

```python
def common_elements(lst1,lst2):
    result=[]
    for i in lst1:
        for j in lst2:
            if i==j:
                result.append(i)
    return result
lst1=[2,3,4,5,6,7,8]
lst2=[4,5,6,7,8,9,10]
print(common_elements(lst1,lst2))
    
```

```python
def merge_sort(lst,low,high):
    mid=(low+high)//2
    if low>=high:
        return 
    r1=merge_sort(lst,low,mid)
    r2=merge_sort(lst,mid+1,high)
    return merging(r1,r2)
def merging(lst5,lst6):
    i=0
    j=0
    result=[]
    while i<len(lst5) and j<len(lst6):
        if lst5[i]<lst6[j]:
            result.append(lst5[i])
            i+=1
        else:
            result.append(lst6[j])
            j+=1
    if i>=len(lst5):
        while j<len(lst6):
            result.append(lst6[j])
            j+=1
    if j>=len(lst6):
        while i<len(lst5):
            result.append(lst5[i])
            i+=1
    return result
lst=[34,11,-5,-67,78,23,45,77,100,-90,9]
low=0
high=len(lst)-1
print(merge_sort(lst,low,high))
```
