```python
def min_max(lst,low,high):
    if low==high:
        min_ele=max_ele=lst[low]
        return min_ele,max_ele
    if high==low+1:
        if lst[low]>lst[high]:
            min_ele=lst[high]
            max_ele=lst[low]
        else:
            min_ele=lst[low]
            max_ele=lst[high]
        return min_ele,max_ele
    mid=(low+high)//2
    left_min,left_max= min_max(lst,low,mid)
    right_min,right_max=min_max(lst,mid+1,high)
    original_min=smaller(left_min,right_min)
    original_max=bigger(left_max,right_max)
    return original_min,original_max
def bigger(a,b):
    if a>b:
        return a
    else:
        return b
def smaller(x,y):
    if x<y:
        return x
    else:
        return y
lst=[12,-5,-6,17,11,5,6,34,-71,99,100,28,-56]
low=0
high=len(lst)-1
print(min_max(lst,low,high))
```

```python
def linear_search(lst,key):
    for i in lst:
        if i==key:
            return True
    return False
lst=[12, -5, -6, 17, 11, 5, 6, 34, -71, 99, 100, 28, -56]
print(linear_search(lst,100))
```

```python
def recursive_bs(lst, low, high, key):
    if low <= high:
        mid = (low + high) // 2
        if key == lst[mid]:
            return True
        elif lst[mid] < key:
            return recursive_bs(lst, mid + 1, high, key)
        else:
            return recursive_bs(lst, low, mid - 1, key)
    return False
lst = [23, 26, 27, 89, 90, 100]
low = 0
high = len(lst) - 1
print(recursive_bs(lst, low, high, 100))  
```
