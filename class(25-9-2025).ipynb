```python
def missing_duplicates(lst):
    i=0
    while i<len(lst):
        if lst[i]!=i+1:
             print("missing:",i+1)
             print("duplicates:",i)
        i+=1
   
        
lst=[1,2,3,3,5]
missing_duplicates(lst)
```

```python
def missing_duplicate(lst):
    for i in lst:
        if lst[i]==i:
            return(f"duplicate:{i} missing:{i+1}")
        
lst=[1,2,3,3,5]
print(missing_duplicate(lst))
```

```python
def two_sum(lst,target):
    for i in range(0,len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i]+lst[j]==target:
                return lst[i],lst[j]
lst=[2,3,4,6,7]
target=5
print(two_sum(lst,target))
    
```

```python
def target_sum(lst,target):
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
          for k in range(j+1,len(lst)):
            if lst[i]+lst[j]+lst[k]==target:
                return (lst[i],lst[j],lst[k])
lst=[1,2,3,4,5]
target=6
print(target_sum(lst,target))
```

```python
def my_discard(s,ele):
    result=[]
    for i in s:
        if i==ele:
            continue
        result.append(i)
    return result    
s={1,2,3,5}
print(my_discard(s,5))
```
